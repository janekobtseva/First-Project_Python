```python
#Aggregating Data with transform() - creating a loyalty flag colum - column containing the maximum frequency of the “order_number”

df_ords_prods_merge_new_variable['max_order'] = df_ords_prods_merge_new_variable.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge_new_variable.head(15)
```

```python
#Deriving Columns with loc()
df_ords_prods_merge_new_variable.loc[df_ords_prods_merge_new_variable['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge_new_variable.loc[(df_ords_prods_merge_new_variable['max_order'] <= 40) & (df_ords_prods_merge_new_variable['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge_new_variable.loc[df_ords_prods_merge_new_variable['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge_new_variable['loyalty_flag'].value_counts(dropna=False)
```

```python
#creating a loyalty flag for existing customers using the transform() and loc() functions.
df_ords_prods_merge_new_variable.loc[df_ords_prods_merge_new_variable['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge_new_variable.loc[(df_ords_prods_merge_new_variable['max_order'] <= 40) & (df_ords_prods_merge_new_variable['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge_new_variable.loc[df_ords_prods_merge_new_variable['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge_new_variable['loyalty_flag'].value_counts(dropna=False)
```

```python
#Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category
df_ords_prods_merge_new_variable.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
#target different types of spenders in their marketing campaigns. 

df_ords_prods_merge_new_variable.groupby('user_id').agg({'prices': ['mean']})
```

```python
df_ords_prods_merge_new_variable['prices_mean'] = df_ords_prods_merge_new_variable.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge_new_variable.head(15)
```

```python
df_ords_prods_merge_new_variable.loc[df_ords_prods_merge_new_variable['prices_mean'] < 10, 'Spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge_new_variable.loc[df_ords_prods_merge_new_variable['prices_mean'] >= 10, 'Spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge_new_variable['Spending_flag'].value_counts(dropna=False)
```

```python
# determine frequent versus non-frequent customers

df_ords_prods_merge_new_variable['days_SPO_median'] = df_ords_prods_merge_new_variable.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge_new_variable.head(15)
```

```python
df_ords_prods_merge_new_variable.loc[df_ords_prods_merge_new_variable['days_SPO_median'] > 20, 'Frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge_new_variable.loc[(df_ords_prods_merge_new_variable['days_SPO_median'] <= 20) & (df_ords_prods_merge_new_variable['days_SPO_median'] > 10), 'Frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge_new_variable.loc[df_ords_prods_merge_new_variable['days_SPO_median'] <= 10, 'Frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge_new_variable['Frequency_flag'].value_counts(dropna=False)
```
